### Multiple plots

In this notebook we will check out how to plot multiple charts in Bokeh

```python
from bokeh.io import output_notebook
from bokeh.plotting import show, figure, gridplot
import numpy as np
```

```python
output_notebook()
```

```python
x1,y1 = np.random.normal(size= 100), np.random.normal(size= 100)
x2,y2 = np.random.normal(size= 100), np.random.normal(size= 100)
x3,y3 = np.random.normal(size= 100), np.random.normal(size= 100)
x4,y4 = np.arange(10), np.random.randint(0,100, size = 10)
```

```python
f1 = figure(width = 200, plot_height = 250, title = 'Circles')
f1.circle(x1,y1, size = 10, color = 'blue', alpha = 0.5)

f2 = figure(width = 200, plot_height = 250, title = 'Triangles')
f2.triangle(x2,y2, size = 10, color = 'red', alpha = 0.5)

f3 = figure(width = 200, plot_height = 250, title = 'Squares')
f3.square(x3,y3, size = 10, color = 'green', alpha = 0.5)

f4 = figure(width = 200, plot_height = 250, title = 'Lines')
f4.line(x4,y4,  color = 'green', alpha = 0.5)


f = gridplot([[f1,f2], [f4,f3]], plot_width = 400, plot_height = 300)
show(f)
```
